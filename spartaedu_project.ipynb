```python
import pandas as pd
import seaborn as sns 
import matplotlib as plt
```

```python
df = pd.read_csv('housing.csv')
df
```

```python
#Number of houses distributed by location (focusing on Bay Area ('near Bay' vs others)

g = sns.relplot(kind="scatter", data=df, x="longitude", y="latitude", hue="ocean_proximity", alpha=0.5, s=20, palette="Set1")
#g.set(title="Geographic Distribution of Houses")
```

```python
# Number of houses by price per location e.g are there more afforable properties on land or by ocean?

g = sns.displot( x="median_house_value", data=df, hue="ocean_proximity", kind="kde", fill=True, alpha=0.6, palette="pastel")
#g.set(title="House Values by Location")
```

```python
# What's the median age by location?

g = sns.boxplot(x="ocean_proximity", y="housing_median_age", data=df, palette="pastel")
#g.set_title("Housing Median Age by Location")
```

```python
# Correlation between age and income by Location.

g = sns.pairplot(vars=["housing_median_age", "median_income"], data=df, hue="ocean_proximity", palette="pastel", height=3, plot_kws={"alpha":0.6})
#g.fig.suptitle("Pair Plot: Housing Age vs Median Income by Location", y=1.02)
```

```python
# Correlation between population and number of households by location.

g = sns.scatterplot(x="population", y="households", data=df, hue="ocean_proximity", alpha=0.6)
#g.set(title="Population vs Households by Location")
```
